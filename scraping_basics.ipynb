```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
## the link to the site
wiki_cc = 'https://en.wikipedia.org/wiki/ISO_3166-1'
## GET request to the site
req = requests.get(wiki_cc)
```

```python
## grab the text from the requested page
## parse it with the HTML parser
soup = BeautifulSoup(req.text, 'html.parser')

table = soup.findAll('tr')

# this is Afghanistan to Zimbabwe
countries = table[1:250]

cc_getter = lambda x: [x[0].get_text(), x[2].get_text()]
```

```python
cc_fields = ['country', 'cc']

cc_dicts = [dict(zip(cc_fields, cc_getter(entry.find_all('td')))) for entry in countries]
countries_df = pd.DataFrame(cc_dicts)
```
